```python
import pandas as pd
import numpy as np
```

```python
movies_md = pd.read_csv("movies_metadata.csv")
movies_keywords = pd.read_csv("keywords.csv")
movies_credits = pd.read_csv("credits.csv")
```

```python
#here .head() shows the 5 rows of the dataset
movies_md.head()
```

```python
#here .info() shows the type of data being used which is now Dataframe
movies_md.info()
```

```python
movies_keywords.head()
```

```python
#here we listed the popular movies whose vote_count is greater than 55
movies_md = movies_md[movies_md['vote_count'] >= 55]
movies_md.head()
```

```python
#Keeping only selected columns
movies_md = movies_md[['id', 'original_title', 'overview', 'genres']]

#here reset_index will do the reindexing/re-numbering of the id of the dataset movies_md
movies_md.reset_index(inplace=True, drop=True)

movies_credits = movies_credits[['id', 'cast']]
```

```python
movies_credits.head()
```

```python
len(movies_md)
```

```python
#converting the id column into int of dataset movies_md
movies_md['id'] = movies_md['id'].astype(int)
```

```python
#merging all three dataset
movies_df = pd.merge(movies_md, movies_keywords, on='id', how='left')
movies_df.reset_index(inplace=True, drop=True)
movies_df = pd.merge(movies_df, movies_credits, on='id', how='left')
movies_df.reset_index(inplace=True, drop=True)

movies_df.head()
```

```python
#Showing the keywords of the first index
movies_df['keywords'][0]
```

```python
#Data Cleaning
ev=[]
def get_genre(i):
    return [x['name'] for x in eval(i)]
```

```python
movies_df['genre']=movies_df['genres'].apply(get_genre)
```

```python
movies_df.head()
```

```python

movies_df.isnull().sum()
```

```python
movies_df.dropna(inplace=True)
movies_df['genre'] = movies_df['genre'].apply(lambda x: ' '.join(x))
movies_df.head()
```

```python
def cast(i):
    return [x['name'] for x in eval(i)]
movies_df['cast']=movies_df['cast'].apply(cast)
```

```python
def keywords(i):
    return [x['name'] for x in eval(i)]
movies_df['keywords']=movies_df['keywords'].apply(keywords)
```

```python
# movies_df['cast'] = movies_df['cast'].apply(lambda x: [i['name'] for i in eval(x)])
# movies_df['keywords'] = movies_df['keywords'].apply(lambda x: [i['name'] for i in eval(x)])
movies_df['cast'] = movies_df['cast'].apply(lambda x: ' '.join(x))
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: ' '.join(x))
```

```python
movies_df.head()
```

```python
#Rearranging the names of the columns
movies_df['tags'] = movies_df['overview']+' '+movies_df['keywords']+' '+movies_df['cast']+' '+movies_df['genre']+' '+movies_df['original_title']
movies_df.head()
```

```python
#Showing the data of first column 'tag'
movies_df['tags'][0]
```

```python
#adding imdb rating column
import random
def imdb(i):
    return random.randint(1,10)
```

```python
movies_df['imdb_rating']=movies_df['id'].apply(imdb)
movies_df.head()
```

```python
movies_df=movies_df.drop('genres',axis='columns')
```

```python
#making questions for 'overview' and create 'ask_overview'
def overview(i):
    return ["overview "+i,'i want to overview '+i,'what happened in '+i,'what is the gist of '+i,'gist of '+i]
movies_df['ask_overview']=movies_df['original_title'].apply(overview)
movies_df.head()
```

```python
#making questions for 'who_acted' in 'actors'
def who_acted(i):
    return ['who acted in '+i, 'who are the actors of '+i, 'who played in '+i]
movies_df['actors']=movies_df['original_title'].apply(who_acted)
movies_df.head()
```

```python
movies_df['actors'][0][0]
```

```python
#making questions for 'genre'
def which_genre(i):
    return ['which genre is '+i,i+' belong to which genre','which genre are contained in '+ i]
movies_df['which_genre']=movies_df['original_title'].apply(which_genre)
movies_df.columns
```
